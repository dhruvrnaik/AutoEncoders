```python
import torch
import numpy as np
from torchvision import datasets
import torchvision.transforms as transforms
```

```python
transform = transforms.Compose([transforms.ToTensor()])
train_data = datasets.MNIST(root="MNIST_data",download=True,train=True,transform=transform)
test_data = datasets.MNIST(root='MNIST_data',download=True,train=False,transform=transforms)
```

```python
batch_size = 20
num_workers = 1

trainloader = torch.utils.data.DataLoader(train_data,batch_size=batch_size,num_workers=num_workers,shuffle=True)
testloader  = torch.utils.data.DataLoader(test_data, batch_size=batch_size,num_workers=num_workers)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
images,_ = next(iter(trainloader))
img=np.squeeze(images[0])
plt.figure(figsize=(15,15))
width,height = img.shape
thresh = img.max()/2.5
img = img.numpy()
for x in range(width):
    for y in range(height):
        val = round(img[x][y],2)
        plt.annotate(str(val),
                     color = 'white' if img[x][y]<thresh else 'black',
                     xy=(y,x),horizontalalignment='center',verticalalignment='center')
        
plt.imshow(img,cmap='gray')
```

```python
import torch.nn as nn

class Autoencoder(nn.Module):
    def __init__(self,enc_dim):
        super(Autoencoder,self).__init__()
        self.enc = nn.Linear(784,enc_dim)
        self.dec = nn.Linear(enc_dim,784)
        self.sig = nn.Sigmoid()
        
    def forward(self,x):
        x = x.view(x.size(0),-1)
        return self.sig(self.dec(self.enc(x)))
        
```

```python
enc_dim=32
model = Autoencoder(enc_dim)
```

```python
criterion = nn.MSELoss()

optimizer = torch.optim.Adam(model.parameters(),lr=0.001)

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
device
```

```python
n_epochs = 20
model = model.to(device)
for epoch in range(1,n_epochs+1):
    train_loss=0.0
    for data,labels in  trainloader:
        
        data = data.view(data.size(0),-1)
        optimizer.zero_grad()
        data = data.to(device)
        output = model(data)
        loss = criterion(output,data)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()   
    train_loss = train_loss/len(trainloader)
    print("Epoch :{} \t Training Loss: {} ".format(epoch,train_loss))
    
```

```python
# obtain one batch of test images
dataiter = iter(trainloader)
images, labels = dataiter.next()

images_flatten = images.view(images.size(0), -1)
# get sample outputs
output = model(images_flatten.cuda())
# prep images for display
images = images.numpy()

# output is resized into a batch of images

output = output.view(batch_size,1,28,28)
# use detach when it's an output that requires_grad
output = output.cpu().detach().numpy()

# plot the first ten input images and then reconstructed images
fig, axes = plt.subplots(nrows=2, ncols=10, sharex=True, sharey=True, figsize=(10,2))

# input images on top row, reconstructions on bottom
for images, row in zip([images, output], axes):
    for img, ax in zip(images, row):
        ax.imshow(np.squeeze(img.transpose((1,2,0))),cmap='gray')
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
```

```python
fin = np.zeros((4*28,4*28))
print(fin.shape)
for i in range(4):
    fin[i*28:i*28+28,0:28] = images[i]
    fin[i*28:i*28+28,28:56] = images[i+4]
    fin[i*28:i*28+28,56:84] = images[i+8]
    fin[i*28:i*28+28,84:112] = images[i+12]
```

```python
plt.figure(figsize=(25,25))
plt.imsave(arr=fin,fname='MNIST_gen_{}'.format(enc_dim),dpi=100,cmap='gray')
plt.imshow(fin,cmap='gray')
```
